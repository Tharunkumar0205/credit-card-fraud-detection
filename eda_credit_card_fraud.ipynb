ðŸ“Š EDA Notes â€“ Credit Card Fraud Detection Dataset

The dataset contains 283,726 transactions after deduplication, with 30 numerical features (Time, V1â€“V28, Amount) and 1 binary target variable (Class).

All features are numeric, making the dataset directly suitable for classical machine learning models without categorical encoding.

No missing values were found across any features, confirming data completeness.

1,081 exact duplicate transactions were identified and removed to prevent data leakage and artificial inflation of model performance.

The target variable is extremely imbalanced, with approximately 0.1667% fraudulent transactions and 99.8333% legitimate transactions.

Due to the severe class imbalance, accuracy is not an appropriate evaluation metric; instead, ROC-AUC and Precisionâ€“Recall AUC will be used.

Features V1â€“V28 are PCA-transformed and already standardized, while Time and Amount require scaling before modeling.

The dataset reflects real-world credit card fraud patterns, where fraudulent transactions are rare and decision thresholds must be tuned carefully.

```python
from sklearn.metrics import precision_recall_curve
import numpy as np
import pandas as pd

precision, recall, thresholds = precision_recall_curve(Y_test, y_proba)

rows = []
for t in thresholds:
    y_pred_t = (y_proba >= t).astype(int)
    
    tp = ((y_pred_t == 1) & (Y_test == 1)).sum()
    fp = ((y_pred_t == 1) & (Y_test == 0)).sum()
    fn = ((y_pred_t == 0) & (Y_test == 1)).sum()
    tn = ((y_pred_t == 0) & (Y_test == 0)).sum()

    precision_t = tp / (tp + fp) if (tp + fp) else 0
    recall_t = tp / (tp + fn) if (tp + fn) else 0

    rows.append([t, precision_t, recall_t, tp, fp, fn, tn])

threshold_df = pd.DataFrame(
    rows, 
    columns=["threshold","precision","recall","TP","FP","FN","TN"]
)

threshold_df.sort_values("precision", ascending=False).head(10)
```

Step 1 â€” Threshold Sweep

```python
import numpy as np
import pandas as pd

thresholds = np.linspace(0.01, 0.99, 60)

rows = []
for t in thresholds:
    y_pred = (calibrated_score >= t).astype(int)

    tp = ((y_pred==1) & (Y_test==1)).sum()
    fp = ((y_pred==1) & (Y_test==0)).sum()
    fn = ((y_pred==0) & (Y_test==1)).sum()
    tn = ((y_pred==0) & (Y_test==0)).sum()

    precision = tp/(tp+fp) if (tp+fp) else 0
    recall = tp/(tp+fn) if (tp+fn) else 0
    f1 = (2*precision*recall)/(precision+recall) if (precision+recall) else 0

    rows.append([t, precision, recall, f1, tp, fp, fn, tn])

df_thr = pd.DataFrame(rows, columns=["t","precision","recall","f1","TP","FP","FN","TN"])
df_thr.sort_values(["precision","recall"], ascending=False).head(10)
```

```python
best = df_thr[df_thr["precision"] >= 0.90] \
            .sort_values("recall", ascending=False) \
            .head(1)

best
```

```python
from sklearn.metrics import confusion_matrix

y_final = (calibrated_score >= best_t).astype(int)

print("Chosen threshold:", best_t)
print(confusion_matrix(Y_test, y_final))
```

1) Make a real prediction function (batch, not 1 row)

```python
import json, joblib
import numpy as np
import pandas as pd

# load artifacts (once)
xgb_raw = joblib.load("models/xgb_fraud_model_raw.joblib")
platt   = joblib.load("models/platt_calibrator.joblib")
best_t  = json.load(open("models/threshold.json"))["threshold"]
cols    = json.load(open("models/feature_columns.json"))["columns"]

def predict_df(df: pd.DataFrame):
    df = df.copy()
    df = df[cols]                          # enforce correct column order
    raw = xgb_raw.predict_proba(df)[:, 1]  # raw prob
    cal = platt.predict_proba(raw.reshape(-1,1))[:, 1]  # calibrated prob
    pred = (cal >= best_t).astype(int)
    out = df.copy()
    out["fraud_proba_raw"] = raw
    out["fraud_proba_cal"] = cal
    out["fraud_pred"] = pred
    return out
```

```python
out = predict_df(X_test.iloc[:20])
out[["fraud_proba_raw","fraud_proba_cal","fraud_pred"]].head()
```

2) Evaluate end-to-end (so you can report metrics)

```python
from sklearn.metrics import classification_report, confusion_matrix

out_all = predict_df(X_test)
y_pred = out_all["fraud_pred"].values

print("Threshold:", best_t)
print(confusion_matrix(Y_test, y_pred))
print(classification_report(Y_test, y_pred, digits=4))
```
